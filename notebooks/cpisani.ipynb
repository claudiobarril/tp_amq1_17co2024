```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Inspecci칩n de Datasets

El descargable que contiene datos de autos contiene 4 archivos con muchos datos, veremos que contiene cada uno:

```python
cars1 = pd.read_csv('../datasets/car data.csv')
cars2 = pd.read_csv('../datasets/CAR DETAILS FROM CAR DEKHO.csv')
cars3 = pd.read_csv('../datasets/Car details v3.csv')
cars4 = pd.read_csv('../datasets/Car details v4.csv')
```

```python
shape = cars1.shape
print(f'[{shape[0]}]observaciones - [{shape[1]}]features')
cars1.describe(include='all').T
```

```python
shape = cars2.shape
print(f'[{shape[0]}]observaciones - [{shape[1]}]features')
cars2.describe(include='all').T
```

```python
shape = cars3.shape
print(f'[{shape[0]}]observaciones - [{shape[1]}]features')
cars3.describe(include='all').T
```

```python
shape = cars4.shape
print(f'[{shape[0]}]observaciones - [{shape[1]}]features')
cars4.describe(include='all').T
```

```python
cars1['Dataset'] = 'cars1'
cars2['Dataset'] = 'cars2'
cars3['Dataset'] = 'cars3'
cars4['Dataset'] = 'cars4'

cars1['anio'] = cars1['Year']
cars2['anio'] = cars2['year']
cars3['anio'] = cars3['year']
cars4['anio'] = cars4['Year']

df_combined = pd.concat([cars1, cars2, cars3, cars4])

plt.figure(figsize=(10, 6))
sns.histplot(data=df_combined, x='anio', hue='Dataset', multiple='stack', shrink=.8)
plt.title('Distribuci칩n de Observaciones por A침o en los 4 Datasets', fontsize=15)
plt.xlabel('A침o')
plt.ylabel('Cantidad de Observaciones')
plt.show()
```

Podemos ver que el data set 3 no solo es el dataset que aporta mayor cantidad de observaciones sino tambi칠n mayor amplitud en el rango de fechas. El dataset 2 aporta buena cantidad de observaciones pero se mantiene en el rango de fechas del 3 y el 4 tiene pocas observaciones pero bastante actuales.

A continuaci칩n graficaremos la distribuci칩n pero en base a los precios. La haremos utilizando solo los datasets 2, 3 y 4 ya que el 1 adem치s de estar en otra escala tiene un aporte marginal debido a las pocas observaciones

```python
cars2['precio'] = cars2['selling_price']
cars3['precio'] = cars3['selling_price']
cars4['precio'] = cars4['Price']

df_combined = pd.concat([cars2, cars3, cars4])

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_combined, x='Dataset', y='precio')
plt.title('Distribuci칩n de Precios', fontsize=15)
plt.xlabel('Dataset')
plt.ylabel('Precio')
plt.show()
```

Se puede notar como los datasets 2 y 3 matienen un mismo rango de precios mientras que el 4 se pasa del umbral con precios mas elevados y heterogeneos.
