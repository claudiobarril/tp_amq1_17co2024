En teor칤a podr칤amos combinar los datasets 2 y 3 ya que su distribuci칩n y su cantidad de features es muy similar. El dataset 2 contiene mileage, engine, max_power, torque y seats que podr칤an estar en realidad relacionadas con el modelo del auto

```python
df_clean = cars3[['name', 'mileage', 'engine', 'max_power', 'torque', 'seats']].dropna()
grouped = cars3.groupby('name')
result = grouped[['mileage', 'engine', 'max_power', 'torque', 'seats']].nunique()
inconsistent_models = result[(result > 1).any(axis=1)]
inconsistent_models
```

```python
models_with_inconsistencies = inconsistent_models.index
inconsistent_rows = df_clean[df_clean['name'].isin(models_with_inconsistencies)]
inconsistent_models = inconsistent_rows.drop_duplicates().sort_values(by='name')
inconsistent_models
```

```python
def get_unique_values(column):
    unique_vals = column.dropna().unique()
    return list(unique_vals)

result = grouped.agg({
    'mileage': lambda x: get_unique_values(x),
    'engine': lambda x: get_unique_values(x),
    'max_power': lambda x: get_unique_values(x),
    'torque': lambda x: get_unique_values(x),
    'seats': lambda x: get_unique_values(x)
})

columns_of_interest = ['mileage', 'engine', 'max_power', 'torque', 'seats']

inconsistent_models = result[~result[columns_of_interest].apply(lambda x: all(len(v) <= 1 for v in x), axis=1)]
inconsistent_models
```

Vemos que en el dataset hay 239 observaciones cuyas caracter칤sticas son inconsistentes entre sus modelos. Dicho de otra forma, no podr칤amos asociar un modelo de auto a un conjunto de caracter칤sticas 칰nicas por lo que si unimos los datasets 2 y 3 muchas tendr칤amos que eliminar las columnas mileage, engine, max_power, torque y seats

Podr칤amos ver si hay duplicidad entre los datasets 2 y 3 

```python
cars2 = cars2.drop(columns=['Dataset'])
cars3 = cars3.drop(columns=['Dataset'])

columns_to_exclude = ['mileage', 'engine', 'max_power', 'torque', 'seats']

cars = [cars2, cars3.drop(columns=columns_to_exclude)] 
combined_cars = pd.concat(cars, ignore_index=True)
duplicates = combined_cars[combined_cars.duplicated(keep=False)]
duplicates
```
